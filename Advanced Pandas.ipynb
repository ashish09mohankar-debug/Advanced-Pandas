2.Category Performance Hierarchy There is a clear hierarchy in product demand across all regions:
Electronics is the top-performing category in every single region.
Accessories consistently holds the middle ground.
Office Supplies is the lowest revenue generator across all regions. This suggests the business's core strength lies in high-value tech hardware rather than general administrative supplies.

3.Regional Specialization in Accessories While the East leads in Electronics, the South and West regions show strong performance in Accessories (both exceeding 64,000). 4.Underperforming Segments & Opportunities The Office category shows the lowest engagement, particularly in the East (14,122) and South (15,504). Given that the East has the highest overall customer traffic, there is a clear strategic opportunity to improve cross-selling tactics to convert Electronics buyers into Office Supply customers.

```python
#Q7 Pivot Table Creation 
#Create a pivot table showing: 
#Rows → Region 
#Columns → Product 
#Values → Total Sales 
#Which region sells the most laptops?
pivot_table = df.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')


display(pivot_table)

top_laptop_region = pivot_table['Laptop'].idxmax()
print(f"The region that sells the most laptops is: {top_laptop_region}")
```

```python
#Q8 Profitability Analysis 
#Calculate the average profit for each product category.
avg_profit = df.groupby("Category")["Profit"].mean().sort_values(ascending = False)
print("Average Profit per Category:")
display(avg_profit)
#Which category is most profitable?
most_profitable_category = avg_profit.idxmax()
print(f"\nThe most profitable category is: {most_profitable_category}")
```

```python
# Q9 Discount Impact Study Analyze the relationship between discount and profit. 
    #Tasks: 
   # 1.Calculate the average profit for each discount level 
   
discount_analysis = df.groupby('Discount')['Profit'].mean().sort_values(ascending=False)

print("Average Profit by Discount Level:")
display(discount_analysis)

# 2.Identify whether higher discounts lead to lower profit
correlation = df['Discount'].corr(df['Profit'])
print(f"\nCorrelation between Discount and Profit: {correlation:.2f}")
```

```python
#Q10 End-to-End Data Wrangling Task Perform the following workflow: 
#Filter the dataset where Sales > 3000 
#Group data by Region and Product 
#Calculate total sales Reshape the result into a pivot table 

high_value_df = df[df['Sales'] > 3000]


wrangled_pivot = high_value_df.pivot_table(index='Region', 
                                           columns='Product', 
                                           values='Sales', 
                                           aggfunc='sum').fillna(0)

print("High-Value Sales Pivot Table (Sales > 3000):")
display(wrangled_pivot)
```
